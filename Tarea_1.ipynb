```python
#CAPITULO 4 - PUNTO 2
```

```python
def funcion(x):
  return (1)/(np.sqrt(1+(np.exp(1)**(-x**2))))
```

```python
x = np.linspace(-10,10,200)
y = funcion(x)
h = 0.05
```

```python
def DerivadaCentral(funcion,x,h):
  d = 0.
  if h != 0:
    d = (funcion(x+h) - funcion(x-h))/(2*h)
  return d

central = DerivadaCentral(funcion,x,h)
```

```python
plt.scatter(x,y)
plt.title("Función")
```

```python
plt.scatter(x,central)
plt.title("Derivada")
```

```python
def DerivadaExacta(x):
  return (x*(np.exp(1)**(-x**2)))/((1+np.exp(1)**(-x**2))**(3/2))
```

```python
error_general = abs(DerivadaCentral(funcion,x,h)-DerivadaExacta(x))
plt.plot(x,error_general)
```
